```python
%matplotlib inline
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from JSAnimation.IPython_display import display_animation
fig = plt.figure()
ax = plt.axes(xlim = (-6, 6), ylim = (-2, 2))
line, = ax.plot([], [], lw = 2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(-6, 6, 101)
    y = 1.75 * np.sin(x - 20*i)
    line.set_data(x, y)
    return line,

    
    
    
    
    
    
    
anim = animation.FuncAnimation(fig, animate, init_func = init, frames =  101, interval = 1, blit = True)
display_animation(anim)
```

```python
fig = plt.figure()
ax = plt.axes(xlim = (-6, 6), ylim = (-2, 2))
line, = ax.plot([], [], lw = 2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(-6, 6, 101)
    y = 1.8 * np.exp(-2*(x - 12*i + 6)**2)
    line.set_data(x, y)
    return line,

    
    
    
    
    
    
    
anim = animation.FuncAnimation(fig, animate, init_func = init, frames =  101, interval = 1, blit = True)
display_animation(anim)
```
